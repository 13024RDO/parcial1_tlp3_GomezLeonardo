### 3- Limpieza de datos

<p>En la primera columna reemplazamos los datos nulos del nombre por no registrado y luego lo mismo con los apellido y provincias no registradas</p>
<p>Se escribieron con mayuscalas los nombres apellidos y provincias</p>
<p>Eliminados los DNI no registrados</p>
<p>Cambiamos los votos que tenian NAN a nulos</p>
<p>Pusimos todos los votos a mayusculas</p>
<p>Cambiamos los valores de dni a tipo de datos INT</p>

```python

df["nombre"] = df["nombre"].fillna("No registrado")
df["apellido"] = df["apellido"].fillna("No registrado")
df["provincia"] = df["provincia"].fillna("No registrado")

df["nombre"] = df["nombre"].str.title()
df["apellido"] = df["apellido"].str.title()
df["provincia"] = df["provincia"].str.title()

df = df.dropna(subset=["dni"])
df = df.dropna(subset=["fecha_votacion"])

df["voto"] = df["voto"].fillna("nulo")
df["voto"] = df["voto"].str.upper()

df["dni"] = df["dni"].astype(int)

```

### 4- Estadisticas Descriptivas

<p>Mostramos como quedaron los datos luego de la limpieza</p>
<p>Mostramos las frecuencia de votos es decir el canto se voto de cada opcion</p>
<p>Mostramos la frecuencia de votos por provincia</p>
<p>Mostramos el porcentaje de votos por cada provincia</p>

```python
df.info()
```

```python
df.describe()
```

```python
frecuencia_votos = df["voto"].value_counts()

print("Frecuencia votos")
print(frecuencia_votos)

print("Frecuencia de votos por provinca")
frecuencia_votos_provincia = df.groupby("provincia")["voto"].value_counts()
print(frecuencia_votos_provincia)

print("Porcentaje de votos por provincia")
porcentaje_votos_provincia = df.groupby("provincia")["voto"].value_counts(normalize=True).unstack(fill_value=0) * 100
print(porcentaje_votos_provincia)

```

### 5 - Graficos de matplolib

```python
import matplotlib.pyplot as plt

frecuencia_votos.plot(kind="bar", stacked=True, figsize=(10,6))

plt.title("Frecuencia de votacion")
```
